```python
le = LabelEncoder()
```

```python
x= final_df.drop('temp',axis = 1)
y=final_df['temp']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
x_train['preciptype'] = le.fit_transform(x_train['preciptype'].astype(str))
x_train['conditions'] = le.fit_transform(x_train['conditions'].astype(str))
x_train['icon'] = le.fit_transform(x_train['icon'].astype(str))
```

```python
scaler = StandardScaler()
```

```python
scaler.fit_transform(x_train)
```

```python
x_test['preciptype'] = x_test['preciptype'].map(lambda s: '<unknown>' if s not in le.classes_ else s)
le.classes_ = np.append(le.classes_, '<unknown>')
```

```python
x_test['conditions'] = x_test['conditions'].map(lambda s: '<unknown>' if s not in le.classes_ else s)
le.classes_ = np.append(le.classes_, '<unknown>')
```

```python
x_test['icon'] = x_test['icon'].map(lambda s: '<unknown>' if s not in le.classes_ else s)
le.classes_ = np.append(le.classes_, '<unknown>')
```

```python
x_test['preciptype'] = le.transform(x_test['preciptype'])
x_test['conditions'] = le.transform(x_test['conditions'])
x_test['icon'] = le.transform(x_test['icon'])
```

```python
scaler.transform(x_test)
```

## Model building 

```python
def predict_model(model , parameters):
    clf = GridSearchCV( model , parameters, n_jobs = -1)
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    train_score = clf.score(x_train,y_train)
    test_score = clf.score(x_test,y_test)
    best_parameters = clf.best_params_
    best_score = clf.best_score_
    
    return train_score, test_score, best_parameters
```

## KNN

```python
knn_reg = KNeighborsRegressor()
knn_parameters = {'n_neighbors':[35,7,9,11,13], 'p':[1, 2]}
```

```python
predict_model(knn_reg , knn_parameters)
```

## Linear Regression

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
y_pred = lin_reg.predict(x_test)
train_score = lin_reg.score(x_train,y_train)
test_score = lin_reg.score(x_test,y_test)



print(train_score, test_score)
```

## SVR

```python
svm_reg = SVR()
svm_parameters = {'kernel':('linear', 'poly', 'rbf', 'sigmoid'), 'C':[1,10]}
```

```python
predict_model(svm_reg ,svm_parameters)
```
